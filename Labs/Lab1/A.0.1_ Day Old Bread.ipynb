Task description:

A bakery sells loaves of bread for $3.49 each. Day-old bread is discounted by 60 percent. Write a program that begins by reading the number of loaves of day-old bread being purchased from the user. Then your program should display the regular price for the bread, the discount because it is a day old, and the total price. All of the values should be displayed using two decimal places, and the decimal points in all of the numbers should be aligned when reasonable values are entered by the user.

```python
print("How many day old breads do you want to buy: ", end ="")
user_input_number_of_day_old_breads = int(input())
print(user_input_number_of_day_old_breads)
number_of_bred = user_input_number_of_day_old_breads
discount_day_old_bread = 0.6 #Convert the percentage discount to a decimal.
ordinary_price = 3.49

sum_without_discount = ordinary_price * user_input_number_of_day_old_breads
sum_total_discount = discount_day_old_bread * sum_without_discount
sum_with_discount = sum_without_discount - sum_total_discount

print('Regular price:\t${0}'.format(round(sum_without_discount,2)))
print("Your discount:\t${0}".format(round(sum_total_discount,2)))
print("Total:\t\t${0}".format(round(sum_with_discount,2)))
```
